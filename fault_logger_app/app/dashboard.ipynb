```python
conn = sqlite3.connect('motors.db')
```

```python
fault_log = pd.read_sql_query("SELECT * FROM fault_log", conn)
fault_log
```

```python
fault = pd.read_sql_query("SELECT * FROM faults", conn)
fault
```

```python
function = pd.read_sql_query("SELECT * FROM function", conn)
function
```

```python
df1 = pd.merge(fault_log, fault, left_on='fault_id', right_on='id')
df1
```

```python
df2 = pd.merge(df1, function, left_on='function_id', right_on='id')
df2['location_id'] = df2['location_id'].astype(int)
df2.info()
```

```python
location= pd.read_sql_query("SELECT * FROM location", conn)
location
```

```python
df3 = pd.merge(df2, location, left_on='location_id', right_on='id')
df3['starter_id'] = df3['starter_id'].astype(int)
df3
```

```python
starter =  pd.read_sql_query("SELECT * FROM starter", conn)
starter
```

```python
df4 = pd.merge(df3, starter, left_on='starter_id', right_on='id')
df4
```

```python
motors = pd.read_sql_query("SELECT * FROM motors", conn)
motors
```

```python
df5 = pd.merge(df4, motors, left_on='motor_id', right_on='id')
df5.columns
```

```python
columns = ['serial_number','number', 'name_x', 'name_y', 'location', 'fault', 'date_created']
```
